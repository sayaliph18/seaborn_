# **Seaborn**

Seaborn is open source,BSD-licensed python library providing high level API for visualizing the data using python programming language.

It is built on top of matplotlib.


**Seaborn Vs Matplotlib**

*It is summarized that if Matplotlib “tries to make easy things easy and hard things possible”, Seaborn tries to make a well-defined set of hard things easy too.”*

*Seaborn helps resolve the two major problems faced by Matplotlib; the problems are −*

1.   Default Matplotlib parameters
2.   Working with data frames



Import libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

# **Histogram using distance plotting**



*   Histograms represent the data distribution by forming bins along the range of
the data and then drawing bars to show the number of observations that fall in each bin.
*   
The sns.histplot function in Seaborn is designed for drawing histograms

*   essential for examining the distribution of continuous data.


*   This function is versatile and allows for extensive customization, making it easier to draw meaningful insights from the data.




```python
#Flexibly plot a univariate distribution of observations.

sns.histplot(data = i['sepal_width'],bins=[2,3,4,5])
```

```python
sns.histplot(data=i)
```

# **Kernel Density Estimation**

Kernel density estimation is a non-parametric way to estimate the distribution of a variable.

A kernel density estimate (KDE) plot is a method for visualizing the distribution of observations in a dataset, analogous to a histogram.

KDE represents the data using a continuous probability density curve in one or more dimensions.

```python
sns.kdeplot(data = i)
```

```python
sns.kdeplot(data = i['petal_width'])
```

#**Pair relationships using pairplot**

To plot multiple pairwise bivariate distributions in a dataset, you can use the pairplot() function.

pairplot works on Numeric data only

Plot pairwise relationships in a dataset.

```python
sns.pairplot(data = i[['sepal_width','petal_width']])
```

```python
sns.pairplot(data=i[['sepal_width','petal_width','species']],hue='species')
```

'hue' is used to visualize the data of different categories in one plot.

 'palette' is used to change the colour of the plot.

```python
sns.pairplot(data=i,hue='species')
```

```python
sns.pairplot(data=i,hue='species',palette='rainbow')
```

# **Boxplot**
- presents five sample statistics - the minimum, the lower quartile, the median,
the upper quartile and the maximum - in a visual display.
-  used to show distributions of numeric data values,
especially when you want to compare them between multiple groups.
- It organizes large amounts of data, and visualizes outlier values.
- used to check statistical information + it detects outliers if any

```python
sns.boxplot(data=i)
```

**Plotting Bivariate Distribution: **

Bivariate Distribution is used to determine the relation between two variables.

# **Joint Plot**

- It gives scatter plot & its strictly bivariate plot

- Draw multiple bivariate plots with univariate marginal distributions.

```python
sns.jointplot(data=i,x = 'sepal_length',y='sepal_width')
```

```python
sns.jointplot(data=i,x ='sepal_length',y='sepal_width',kind='hex')
```

```python
sns.jointplot(data=i,x = 'sepal_length',y='sepal_width',kind='reg')
```
