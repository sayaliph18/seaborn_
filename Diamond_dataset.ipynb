```python
#Import libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Load dataset
d = sns.load_dataset('diamonds')
d.head()
```

```python
sns.histplot(data=d['carat'])
```

```python
sns.histplot(data=d['carat'],bins=[0,1,2,3,4])
```

```python
sns.jointplot(data =d,x = 'x',y='y')
```

```python
sns.jointplot(data =d,x = 'x',y='z',kind='hex')
```

```python
sns.jointplot(data =d,x = 'x',y='z',hue='cut')
```

```python
sns.jointplot(data =d,x = 'x',y='z',hue='color')
```

```python
sns.kdeplot(data=d)
```

```python
sns.kdeplot(data = d[['carat','depth']])
```

```python
sns.kdeplot(data = d[['carat']])
```

```python
sns.pairplot(data=d)
```

```python
sns.pairplot(data=d[['carat','cut','color','clarity','depth','table','price']])
```

```python
sns.pairplot(data=d[['carat','cut','color','clarity','depth','table','price']],hue='cut')
```

```python
sns.pairplot(data=d[['carat','cut','color','clarity','depth','table','price']],hue='cut',palette='rainbow')
```

```python
sns.boxplot(data = d)
```
