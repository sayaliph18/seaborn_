**Categorical Plotting**

When one or both the variables under study are categorical, we use plots like stripplot(), swarmplot(), etc,.

```python
#Load libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Load penguin dataset
p = sns.load_dataset('penguins')
p.head()
```

```python
# check unique species
p.species.unique()
```

```python
# check unique island
p.island.unique()
```

```python
# check unique sex
p.sex.unique()
```

```python
p.info()
```

```python
#check missing values
p.isna().sum()
```

```python
# so we can remove nan values as its % is low
#p.dropna() # this is temp in order to make it permanent use inplace=True
p.dropna(inplace=True)
```

```python
p.shape
```

```python
p.isna().sum()
```

```python
#check columns
p.columns
```


#stripplot()
- Categorical Scatter plots
- stripplot() is used when one of the variable under study is categorical.
- It represents the data in sorted order along any one of the axis.
-Draw a scatterplot where one variable is categorical.


```python
sns.stripplot(data = p,x='species',y='body_mass_g')
```

```python
# in above plot i want to see genderwise distribustion
sns.stripplot(data = p,x='species',y='body_mass_g',hue='sex')
```

From 3 categories: Male penguins have more body mass than female penguins and Gentoo Penguins have highest body mass among all.

```python
sns.stripplot(x='species',y='body_mass_g',data=p,hue='island')
```

Performing comparison within categories.

## Box Plots
Boxplot is a convenient way to visualize the distribution of data through their quartiles.

```python
sns.boxplot(data=p['body_mass_g'])
```

```python
sns.boxplot(data=p)
```

```python
sns.boxplot(data=p[['bill_length_mm','bill_depth_mm']])
```

```python
sns.boxplot(data=p[['bill_length_mm','bill_depth_mm']])
plt.title('Sample boxplot')
plt.xlabel('Bill Infomation')
plt.ylabel('stats')
```

```python
# assign labels to plot
ax = sns.boxplot(data=p[['bill_length_mm','bill_depth_mm']])
ax.set_title('Sample BOX plot')
ax.set_xlabel('Bill Info')
ax.set_ylabel('Stat info')
```

```python
sns.boxplot(x='species',y='body_mass_g',data=p,hue='sex')
```

## Bar Plot
- The barplot() shows the relation between a categorical variable and a continuous variable.
- The data is represented in rectangular bars where the length the bar represents the proportion of the data in that category.

```python
sns.barplot(x='species',y = 'island',data=p)
```

```python
#load titanic data
t = sns.load_dataset('titanic')
t.head(2)
```

```python
# check survival rate with respect to gender
sns.barplot(x='sex',y='survived',data=t)
```

```python
#hue=class
sns.barplot(x='sex',y='survived',data=t,hue='class')
```

```python
# embarked vs survived
sns.barplot(x='embarked',y='survived',data=t)
```

## count plot
- A special case in barplot is to show the no of observations in each category rather than
- computing a statistic for a second variable. For this, we use countplot()

```python
sns.countplot(data =t,x = 'class',hue = 'survived')
```

```python
sns.countplot(data=t,x='sex',hue='survived')
```

```python
sns.countplot(data=p,x='species',palette='Blues')
```

#Violine plot
kde+boxplot

A violin plot plays a similar role as a box-and-whisker plot.

It shows the distribution of data points after grouping by one (or more) variables.

Unlike a box plot, each violin is drawn using a kernel density estimate of the underlying distribution.

```python
# load tips dataset
tp = sns.load_dataset('tips')
tp.head(3)
```

```python
sns.violinplot(data=tp.total_bill)
```

```python
sns.violinplot(x='sex',y='total_bill',data=tp)
```

```python
sns.violinplot(x='day',y='tip',data=tp,hue='smoker',palette='Reds')
```

```python
sns.histplot(x='total_bill',data=tp,palette='kkk')
```
